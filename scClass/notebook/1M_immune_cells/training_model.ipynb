```python
matrix,label = scClass.read_input(matrix=path_matrix,label=path_label)
```

## loading dataset...

```python
pred_y = scClass.predict(dataset.test_x,model)
```

```python
scClass.confusion_matrix(y_pred=pred_y,y_true=dataset.test_y)
```
