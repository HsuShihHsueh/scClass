```python
#  python script
import sys; sys.path.append("/home/shhsu/download/scClass")
import scClass
from scClass.main import Model
```

```python
def dataset(file):
    return"../../../../scClass_dataset/1M_immune_cells/"+file

path_matrix = dataset('data_train.npz') 
path_label  = dataset('data_train_v2.csv')
```

## AUTO RUN

```python
scClass.auto_run(load_model=True,save=False,matrix=path_matrix,label=path_label)
```

## STEP

```python
# scClass.main.EPOCH = 10  #it can change hyperparameter 
```

```python
matrix,label = scClass.read_input(matrix=path_matrix,label=path_label)
```

```python
matrix,label = scClass.random(matrix,label,seed=0)
```

```python
dataset = scClass.npz_dataloader(matrix,label)
```

```python
model = scClass.Model()
```

```python
scClass.training(dataset,model)
```

```python
y_pred = scClass.predict_batch(dataset.test_x,model)
```

```python
scClass.confusion_matrix(y_pred=y_pred,y_true=dataset.test_y)
```
